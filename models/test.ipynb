```python
import requests
import json

def generate_json(prompt, model="gemma2:2b"):
    url = "http://localhost:11434/api/generate"

    full_prompt = f"""
You must respond ONLY in JSON format with keys:
category, department, summary, date, priority, status, details.

Do not include explanations.

User query:
{prompt}
"""

    data = {
        "model": model,
        "prompt": full_prompt,
        "stream": False
    }

    response = requests.post(url, json=data)
    return response.json()   # Will contain the JSON string from model

result = generate_json("Generate a ticket for server down")
final_data = json.loads(result["response"])
print(final_data)

```
